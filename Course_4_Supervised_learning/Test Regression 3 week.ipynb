```python
idlnchld_noans=list()
heduc_noans=list()
usemeth_noans=list()
hed=data['heduc']
hed=np.array(hed)
idl=data['idlnchld']
idl=np.array(idl)
use=data['usemeth']
use=np.array(use)
for i in range(len(hed)):
    if hed[i]!=hed[i]:
        heduc_noans.append(1)
        hed[i] = -2
    else:
        heduc_noans.append(0)
    if idl[i]!=idl[i]:
        idlnchld_noans.append(1)
        idl[i] = -1
    else:
        idlnchld_noans.append(0)
    if use[i]!=use[i]:
        usemeth_noans.append(1)
        use[i] = -1
    else:
        usemeth_noans.append(0)
```

```python
data['usemeth']=use
data['heduc']=hed
data['idlnchld']=idl
data['idlnchld_noans']=idlnchld_noans
data['heduc_noans']=heduc_noans
data['usemeth_noans']=usemeth_noans
```

```python
import statsmodels

import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

```python
m1 = smf.ols('ceb ~ age+educ+religion+idlnchld+knowmeth+usemeth+agefm+heduc+urban+electric+radio+tv+bicycle+nevermarr+idlnchld_noans+heduc_noans+usemeth_noans', data=data)
fitted = m1.fit()
fitted.summary()
```

```python
print('Breusch-Pagan test: p=%f' % sms.het_breushpagan(fitted.resid, fitted.model.exog)[1])
```

```python
statsmodels.stats.diagnostic.het_breuschpagan(fitted.resid, fitted.model.exog)[1]
```

```python
m2 = smf.ols('ceb ~ age+educ+religion+idlnchld+knowmeth+usemeth+agefm+heduc+urban+electric+radio+tv+bicycle+nevermarr+idlnchld_noans+heduc_noans+usemeth_noans', data=data)
fitted = m2.fit(cov_type='HC1')
fitted.summary()
```

```python
m3 = smf.ols('ceb ~ age+educ+idlnchld+knowmeth+usemeth+agefm+heduc+urban+electric+bicycle+nevermarr+idlnchld_noans+heduc_noans+usemeth_noans', data=data)
fitted = m3.fit()
fitted.summary()
```

```python
statsmodels.stats.diagnostic.het_breuschpagan(fitted.resid, fitted.model.exog)[1]
```

```python
m4 = smf.ols('ceb ~ age+educ+idlnchld+knowmeth+usemeth+agefm+heduc+urban+electric+bicycle+nevermarr+idlnchld_noans+heduc_noans+usemeth_noans', data=data)

fitted = m4.fit(cov_type='HC1')
fitted.summary()
```

```python
print ("F=%f, p=%f, k1=%f" % m2.fit().compare_f_test(m4.fit()))
```
