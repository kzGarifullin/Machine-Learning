```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
# зададим нормально распределенную случайную величину
expon_rv = sts.expon()# экспоненциальное распределение f(x)=exp(-x)
sample=expon_rv.rvs(size=1000)# сгенерируем 1000 значений
x = np.linspace(0,15,1000)#Функция linspace() возвращает одномерный массив из указанного количества элементов, значения 
#которых равномерно распределенны внутри заданного интервала.
# посчитаем теоритеческую плотность вероятности для нашего распределения (pdf - probability density function)
#cumulative density function- cdf
pdf = expon_rv.pdf(x) #вычисляем значение функции плотности вероятности нормального распределения на x
plt.plot(x, pdf)
plt.hist(sample, normed=True)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
d=[5,10,30, 50]
for i in d:
    n=i #объем нашей выборки
    mu = 1.0 # из википедии известно E[x]=1/f=1; D[x]=1/f^2=1 для f(x)=f*exp(-fx) в нашем случае f=1
    # Пусть E[x]-матожидание, а D[x]-дисперсия
    # По центральной предельной теореме выборочные средние приближаются нормальным распределением N(E[x],D[x]/n) 
    # Тогда в нашем случае они приближаются N(μ=1,σ^2=1/n)
    sigma = (1./n)**0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)# зададим нормально распределенную случайную величину
    x = np.linspace(0,15,1000)
    pdf = norm_rv.pdf(x)
    arr=list()#массив из выборочных средних по выборкам объема n
    for i in range(1000):
        sample = expon_rv.rvs(size=n)#генерируем выборку
        s=np.mean(sample)#среднее значение массива
        arr.append(s)#добавим в массив 
    plt.plot(x, pdf,linewidth = 2.5)#строим график
    plt.hist(arr, normed=True)#строим гистограмму
    plt.ylabel('$F(x)$')
    plt.xlabel('$x$')
    plt.title('n = %d' % n)
    plt.show()
```

Вывод: как видно из гистограмм чем больше n, тем лучше нормальное распределение приближает распределение выборочных средних.
