```python
from scipy import linalg
from matplotlib import pylab as plt
import numpy as np
from math import sin
from math import exp
def f(x): 
    k = sin(x / 5.) * exp(x / 10.) + 5 * exp(-x / 2.)
    return k
f1=f(1.0)
f2=f(15.0)
f3=f(4.0)
f4=f(10.0)
a=np.array([[1., 1., 1., 1.], [15.**3,15.**2, 15., 1.],[4.**3, 4.**2, 4., 1.], [1000.,100.,10.,1.]])
b=np.array([f1, f2, f3, f4])
sol = linalg.solve(a,b)
print sol
x = np.arange(1.,15.,0.1)
y=np.sin(x / 5.) *np.exp(x / 10.) + 5 *np.exp(-x / 2.)
g=x*x*x*sol[0]+x*x*sol[1]+x*sol[2]+ sol[3]
plt.plot(x,y)
plt.plot(x,g)
plt.show()
```
